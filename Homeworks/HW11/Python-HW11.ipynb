```python
t=fx-gx
```

```python
t
```

```python
x1,x2,x3,x4=solve(t)
```

```python
solve(t)
```

```python
x1=x1.evalf(chop=True)
x2=x2.evalf(chop=True)
x3=x3.evalf(chop=True)
x4=x4.evalf(chop=True)
```

```python
x1, x2, x3, x4
```

```python
y1=fx.subs(x,x1)
y2=fx.subs(x,x2)
y3=fx.subs(x,x3)
y4=fx.subs(x,x4)
```
