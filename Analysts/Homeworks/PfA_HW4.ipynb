```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Задача 1
- Постройте график
- Назовите график
- Сделайте именование оси x и оси y
- Сделайте выводы

```python
df = pd.read_csv('kc_house_data.csv')
df.head()
```

### 1.2 Изучите стоимости недвижимости

```python
plt.figure(figsize=(20, 8))
sns.histplot(df['price'])
plt.ticklabel_format(style='plain')
plt.title('Распределение стоимости недвижимости')
plt.xlabel('Стоимость')
plt.ylabel('Кол-во');
```

### 1.3 Изучите распределение квадратуры жилой

```python
plt.figure(figsize=(20, 6))
sns.histplot(df['sqft_living'])
plt.ticklabel_format(style='plain')
plt.title('Распределение жилой площади')
plt.xlabel('Площадь')
plt.ylabel('Кол-во');
```

### 1.4 Изучите распределение года постройки

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['yr_built'])
plt.ticklabel_format(style='plain')
plt.title('Распределение года постройки')
plt.xlabel('Год постройки')
plt.ylabel('Кол-во');
```

## 3 Задача

### 2.1 Изучите распределение домов от наличия вида на набережную
Постройте график

```python
data = df['waterfront'].value_counts()
data
```

```python
labels = ['С видом на набережную', 'Без вида']
plt.pie(data, autopct='%1.1f%%', labels=labels)
plt.title('Распределение домов от наличия вида на набережную');
```

### 2.2 Изучите распределение этажей домов

```python
data = df['floors'].value_counts()
data
```

```python
plt.figure(figsize=(6, 4))
sns.barplot(x = data.index, y = data)
plt.title('Этажность домов')
plt.xlabel('Кол-во этажей')
plt.ylabel('Кол-во')
plt.xticks();
```

### 2.2 Изучите распределение состояния домов

```python
data = df['condition'].value_counts()
data
```

```python
plt.figure(figsize=(6, 4))
sns.barplot(x = data.index, y = data)
plt.title('Состояние домов')
plt.xlabel('Состояние дома')
plt.ylabel('Кол-во');
```
